```python
# chargement du fichier
%matplotlib inline 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importation du dataset

```python
data = pd.read_excel("mobilityref-france-irve-220.xlsx")
data.head()
```

```python
data.shape
```

```python
data.describe(include="all")
```

```python
data=data.dropna(axis="columns").head()
```

```python
data.shape
```

```python
data.head()
```

```python
# Calculer la matrice de corrélation
correlation_matrix = data.corr()
```

```python
# Paramètres de la visualisation
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation")

# Afficher la visualisation
plt.show()
```

Boîtes à moustaches pour les 63 colonnes

```python
# Paramètres de la visualisation
plt.figure(figsize=(15, 8))
sns.boxplot(data, orient="h", palette="Set2")

# Ajouter un titre à la visualisation
plt.title("Boîtes à moustaches pour les 63 colonnes")

# Afficher la visualisation
plt.show()
```

```python
data['longitude'] = pd.to_numeric(data['longitude'], errors='coerce')
data['latitude'] = pd.to_numeric(data['latitude'], errors='coerce')

# Paramètres de la visualisation
plt.figure(figsize=(10, 6))

# Créer un graphique de dispersion en utilisant les colonnes de longitude et de latitude avec seaborn
sns.scatterplot(x='longitude', y='latitude', data=data, alpha=0.5)

# Ajouter des titres et des étiquettes
plt.title('Graphique de Dispersion Longitude vs Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Afficher la visualisation
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Créer une colonne géométrique à partir de longitude et latitude
geometry = [Point(xy) for xy in zip(data['longitude'], data['latitude'])]
gdf = gpd.GeoDataFrame(data, geometry=geometry)
```

```python
# Paramètres de la visualisation
fig, ax = plt.subplots(figsize=(10, 6))

# Tracer la carte du monde
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(ax=ax, color='lightgrey')

# Tracer vos données sur la carte
gdf.plot(ax=ax, color='blue', markersize=10)

# Ajouter des titres
plt.title('Carte du Monde avec les Données')

# Afficher la visualisation
plt.show()
```
